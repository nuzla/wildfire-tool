```
import pandas as pd
import numpy as np
from google.colab import files
import haversine as hs
from haversine import Unit
```

```
dfv2 = pd.read_excel('https://drive.google.com/uc?id=1G8Fv_U74xFL2hn0OENuB9l11RvsjydbZ', header=0)
```

```
dfv2.shape
```

```
dfv2.head()
```

```
df_nc_h28v11 = pd.read_excel('https://drive.google.com/uc?id=1wkb6fpMAHE_yDf2ip0RoFgHIkS_rAqug', header=0)
```

```
df_nc_h28v11=df_nc_h28v11[(df_nc_h28v11['lfmc_mean'].notnull()) & (df_nc_h28v11['lfmc_stdv'].notnull())]
```

```
df_nc_h28v11.head()
```

```
df_nc_h27v12 = pd.read_excel('https://drive.google.com/uc?id=1rb2xt5Z6Sofk0JD8AV6k7yUccKK0KFHJ', header=0)
```

```
df_nc_h27v12=df_nc_h27v12[(df_nc_h27v12['lfmc_mean'].notnull()) & (df_nc_h27v12['lfmc_stdv'].notnull())]
```

```
df_nc_h27v12.head()
```

```
df_nc_h29v11 = pd.read_excel('https://drive.google.com/uc?id=1r7riXH0U5CGiiJKEkQCzi3Puws7TLCdf', header=0)
```

```
df_nc_h29v11=df_nc_h29v11[(df_nc_h29v11['lfmc_mean'].notnull()) & (df_nc_h29v11['lfmc_stdv'].notnull())]
```

```
df_nc_h29v11.head()
```

```
df_nc=pd.concat([df_nc_h28v11, df_nc_h27v12, df_nc_h29v11], ignore_index=True)
```

```
df_nc.head()
```

```
df_nc.shape
```

```
df_elevation = pd.read_excel('https://drive.google.com/uc?id=1sLKSdeL0mXrYPGD0UnR8SGVxODi1kDK2', header=0)
```

```
df_elevation.shape
```

```
df_elevation.head()
```

```
df_m1=pd.merge(dfv2, df_nc, how="inner", on=['fih_year1','fih_date1','LAT','LON','MODIS'])
```

```
df_m1.shape
```

```
df_m1.head()
```

```
dfv3=pd.merge(df_m1, df_elevation, how="inner", on=['LAT','LON','MODIS'])
```

```
dfv3.shape
```

```
dfv3['dis_latlon_n_nclatlon']=np.nan
```

```
for index, row in dfv3.iterrows():
  dfv3.at[index, 'dis_latlon_n_nclatlon']=hs.haversine((row['LAT'],row['LON']),(row['NC_LAT'],row['NC_LON']),unit=Unit.METERS)
```

```
dfv3.sample(20)
```

```
dfv3.describe()
```
