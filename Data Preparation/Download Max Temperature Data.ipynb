```
import pandas as pd
from google.colab import files
import os
```

```
dates = pd.DataFrame(pd.date_range(start='1/1/2000', end='31/12/2020'), columns=['date'])
```

```
outfile='MaxTemperatureWesterAustralia.grid'
```

```
dates.shape
```

```
os.system('> '+outfile)
for index, row in dates.iterrows():
  try:
    d=row['date'].strftime('%Y%m%d')
    filename=d+d+'.grid'
    if (index%30)==0:
      print('Processing : '+d)
    r1=os.system('wget "http://www.bom.gov.au/web03/ncc/www/awap/temperature/maxave/daily/grid/0.05/history/nat/'+filename+'.Z"')
    if r1==0:
      r2=os.system('uncompress '+filename+'.Z')
      if r2==0:
        r3=os.system(" grep '^ ' "+filename+" | sed 's/^ //' |  awk '{print \""+d+"\",NR-1,$0}' >> "+outfile)
    else:
      print("Download failed : "+filename+'.Z')
    os.remove(filename)
  except Exception as e:
      print(e)
```
