```
import pandas as pd
import requests
import time
import numpy as np
from google.colab import files
```

```
df = pd.read_excel('https://drive.google.com/uc?id=19mtqIjQvbJhRGZ0A7aYTFPOySkGMbds0', header=0)
df.head()
```

```
df['ELEVATION']=np.nan
```

```
i=0
q="";
for index, row in df.iterrows():
  i=i+1
  q=q+str(row['LAT'])+','+str(row['LON'])+'|'
  if i==100:
    i=0
    print("Index : "+str(index))
    try:
      response = requests.get("https://api.opentopodata.org/v1/etopo1?locations="+q)
      print(response.status_code)
      if response.status_code==200:
        data=response.json()
        print("Status : "+data['status'])
        #print(response.text)
        if data['status']=='OK':
          k=0;
          for j in range(index-99, index+1):
            #print(data['results'][k]['elevation'])
            df.at[j, 'ELEVATION'] = data['results'][k]['elevation']
            k=k+1
      else:
        print(response.text)
      time.sleep(5)
      q="";
    except Exception as e:
      print(e)
```

```
df.to_excel('WesternAustraliaFireDataUniqLatLonElevation.xlsx',index=False)
files.download('/content/WesternAustraliaFireDataUniqLatLonElevation.xlsx')
```
