```
temp[['Station Number','Year', 'Month', 'Day', 'Average daily temperature (calculated by averaging the max and min temperatures) in Degrees C', \
      'Average daily air temperature (using all available observations) in Degrees C','Average daily dew point temperature in Degrees C','Speed of maximum wind gust in km/h', \
      'Mean daily wind speed in km/h','Mean daily wind speed in km/h','Average daily mean sea level pressure in hPa','Average daily station level pressure in hPa']].head()
```

```
temp=temp.rename(columns={"Station Number": "station_number", "Average daily temperature (calculated by averaging the max and min temperatures) in Degrees C": "station_mean_temperature_c",\
                     "Average daily air temperature (using all available observations) in Degrees C":"station_air_temperature_c","Average daily dew point temperature in Degrees C":"Station_TDMEAN_c",\
                     "Average daily mean sea level pressure in hPa":"Mean_Sea_Level_Pressure","Average daily station level pressure in hPa":"Mean_Station_Pressure",\
                     "Mean daily wind speed in km/h":"Mean_Wind_Speed", "Speed of maximum wind gust in km/h":"Maximum_Sustained_Wind_Speed"})
```

```
temp['date']=temp['Year'].astype(str)+"/"+temp['Month'].astype(str)+"/"+temp['Day'].astype(str)
```

```
temp
```

```
df=temp[['station_number','date','station_mean_temperature_c','station_air_temperature_c','Station_TDMEAN_c','Maximum_Sustained_Wind_Speed','Mean_Wind_Speed','Mean_Sea_Level_Pressure','Mean_Station_Pressure']]
```

```
df.sample(100)
```

```
df.to_csv('WAWeatherStationData.csv',index=False)
```
